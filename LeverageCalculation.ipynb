```python
# Import packages
import yfinance as yf
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.stattools as ts
from matplotlib import pyplot as plt
import statistics
import math
```

```python
#target volatility
target_vol = 0.125
```

```python
#download data for close
fut_list = ['SPY']
data2= yf.download(fut_list, start="2018-01-01", end="2023-03-14")['Adj Close'] 
df2 = pd.DataFrame(data=data2)
```

```python
#volatility calculation
vol = pd.DataFrame(data=(((df2.pct_change(1).rolling(20)).std()))*(252**0.5))
vol
```

```python
bob = target_vol/vol
np.clip(bob, 0, 1).plot()
plt.title('Leverage(12.5% Vol Target)')
plt.xlabel('Date')
plt.ylabel('Leverage')
```
